<blockquote><h3><b>The Problem Statement</b></h3>
Currently Rental bikes are introduced in many urban cities for the enhancement of mobility comfort. It is important to make the rental bike available and accessible to the public at the right time as it lessens the waiting time. Eventually, providing the city with a stable supply of rental bikes becomes a major concern. The crucial part is the prediction of bike count required at each hour for the stable supply of rental bikes. 
<br>
<br>
The challnge/task for this project is to estimate the number of bikes needed at a particular hour on a given day of the year so that the number of bikes available for rental is as close to the the actual demand as possible. 
</blockquote>


<blockquote><h3><b>The Dataset</b></h3>
The dataset contains weather information (Temperature, Humidity, Windspeed, Visibility, Dewpoint, Solar radiation, Snowfall, Rainfall), the number of bikes rented per hour and date information.<br><br>
<h4>Attribute Information:<h4>
<ul>
  <li>Date : Date, Month and Year</li>
  <li>Rented Bike count - Count of bikes rented at each hour</li>
  <li>Hour - Hour of he day between 0-23</li>

  <li>Humidity - Humidity at a given Hour in %</li>
  <li>Windspeed - Wind speed at the given Hour in m/s</li>
  <li>Visibility - Visisbility at the given Hour</li>
  <li>Dew point temperature - Dew Point Temprature in 째Celsius</li>
  <li>Solar radiation - Amount of Solar radiation at the given hour MJ/m2</li>
  <li>Rainfall - Rainfall at the given hour in mm</li>
  <li>Snowfall - Snowfall at the given hour in cm</li>
  <li>Seasons - Winter, Spring, Summer, Autumn</li>
  <li>Holiday - Holiday or No holiday</li>
  <li>Functional Day - NoFunc(Non Functional Hours), Fun(Functional hours)</li>
</ul>

```python
# Defining upper Bound
IQR = np.quantile(data["Rented Bike Count"], 0.75) - np.quantile(data["Rented Bike Count"], 0.25)
upper_bound = np.quantile(data["Rented Bike Count"], 0.75) + (1.5 * IQR)
print(upper_bound)
```

```python
# Looking at rows above upper bound
data[data["Rented Bike Count"] > upper_bound].shape

# There are 152 rows where the rented bike  count exceeds upper bound 152 rows is approximately 1.80% hence we drop the outliers
```

```python
# dropping rows where Rented Bike Count is more than the upper bound
data = data[data["Rented Bike Count"] < upper_bound]

# Plotting to see if there are still any outliers
sns.boxplot(data["Rented Bike Count"])
```

<blockquote>There are few points but they are very close to the upper Bound and hence we keep those</blockquote>

<h1>Handling the Date Column</h1>
<ol>
  <li>The day of the week will be extracted to make a new column as integer values between 0 and 6 to feed into the model</li>
  <li>The month number will be extracted from the date column as an integer value to feed into the models</li>
</ol>

```python
# Taking day of the week as an ordinal feature to see how the day of the week influences bike rentals

data["Day of Week"] = data["Date"].apply(lambda x : x.weekday())
data.head()

# Year doesnt seem relevant as there are only 2 years in the whole dataset
```

```python
# Converting dates to useful features.
# 1. Taking the month of the year as an integer from the date column
data["Month"] = data["Date"].apply(lambda x : x.month)
data.drop("Date", axis=1, inplace=True)
data.head()
```

<blockquote>Finally encoding the Seasons and Holiday</blockquote>

```python
# Encoding Holiday and Seasons to discrete integers
enc = OrdinalEncoder()
enc.fit(data[["Holiday","Seasons"]])
data[["Holiday", "Seasons"]] = enc.transform(data[["Holiday","Seasons"]])
```

<blockquote>Now our data </blockquote>

```python
# Fitting and Testing a Baseline Model without any preprocessing of the data
# Building the first simple model
lr_model = LinearRegression()
X = data[[col for col in data.columns if col!= "Rented Bike Count"]]
y = data[["Rented Bike Count"]]
```

```python
# Splitting the Data

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=111)
```

```python
# Fitting the Data
lr_model.fit(x_train,y_train)
```

```python
# Storing predictions in an array
predictions_lr_model = lr_model.predict(x_train)
```

```python
# Checking the R squared score of the Model
lr_model.score(x_test, y_test)
```

<h2>The R suared score is very low at 0.5620</h2>
<h2>The Target feature is positively skewed so it should be transformed using the squre root tranformation to correct the skewness.</h2>

```python
# Transforming the target variable to eliminate the skewness in the data
data["Rented Bike Count"] = np.sqrt(data["Rented Bike Count"])
```

```python
# Chicking the distribution of transformed column
sns.distplot(data["Rented Bike Count"])
```

<blockquote><h2>The distribution of Rented Bike Count is less skewed than before.</h2></blockquote>

```python
# Checking metrics for linear model
predictions_lr_model_test = lr_model.predict(x_test)
predictions_lr_model_train = lr_model.predict(x_train)
```

```python
# Checking Metrics on Training set
print(f"Mean Absolute Error for Training Set is : {mean_absolute_error(y_train, predictions_lr_model_train)}")
print(f"Mean Squared Error for Training Set is : {mean_squared_error(y_train, predictions_lr_model_train)}")
print(f"R squared score for Training Set is: {lr_model.score(x_train, y_train)}")

print("__"*25 + "\n")

# Checking Metrics on Test Set
print(f"Mean Absolute Error for Test Set is: {mean_absolute_error(y_test, predictions_lr_model_test)}")
print(f"Mean Squared Error for Test Set is : {mean_squared_error(y_test, predictions_lr_model_test)}")
print(f"R squared score for Test Set is: {lr_model.score(x_test, y_test)}")
```

<blockquote><h2>A simple linear regression model has a MAE of 5.30 MSE of  46.78 and R2 score of 0.63</h2>

<h2>The Linear Model, however, is not able to model the complexity present in the dataset. XGBRegressor or Random Forest Regressor might be a better model for the problemat hand.</h2></blockquote>

```python
# Creating XGBRegressor Model
xg_model = XGBRegressor(random_state=123, objective='reg:squarederror')
```

```python
# Fitting the data
xg_model.fit(x_train, y_train)
```

```python
# Predicting Results
predictions_xg_model_test = xg_model.predict(x_test)
predictions_xg_model_train = xg_model.predict(x_train)
```

```python
# Checking Metrics on Training set
print(f"Mean Absolute Error for the Training Set is : {mean_absolute_error(y_train, predictions_lr_model_train)}")
print(f"Mean Squared Error for the Training Set is : {mean_squared_error(y_train, predictions_lr_model_train)}")
print(f"R squared score For the Training Set is : {xg_model.score(x_train, y_train)}")

print("__"*25 + "\n")

# Checking Metrics on Test Set
print(f"Mean Absolute Error for the Test Set is: {mean_absolute_error(y_test, predictions_xg_model_test)}")
print(f"Mean Squared Error for the Test Set is : {mean_squared_error(y_test, predictions_xg_model_test)}")
print(f"R squared score for the Test Set is: {xg_model.score(x_test, y_test)}")
```

<h3>The next natural step is to use GridSearchCV to tune the hyperparameters and cross validate so that our model's performace is better judged.</h3>
</blockquote>

```python
# lets try GridSearchCV for hyper-parameter tuning and cross validation to see if the model scores improves further.
params_grid = {
    # Max Depth of the tree before the tree terminates in a leaf node.
    'max_depth' : [5,6,7],
    # Number of estimators or descion trees to train.
    'n_estimators' : [80,90,100],
    # L1 or Lasso regularization parameter.
    'reg_alpha' : [3,4,5],
    # L2 or Ridge regularization parameter.
    'reg_lambda' : [3,4,5],
}
```

```python
# Making a GridSearchCV instance

xgb_cv = GridSearchCV(XGBRegressor(random_state=123, objective='reg:squarederror'),param_grid=params_grid)
```

```python
# Training the Model.
xgb_cv.fit(x_train, y_train)
```

```python
# Predicting Results
predictions_xg_model_test_cv = xgb_cv.predict(x_test)
predictions_xg_model_train_cv = xgb_cv.predict(x_train)
```

```python
# Getting the best parameters from the grid search of hyperparameters.
xgb_cv.best_estimator_
```

```python
# Checking Metrics on Training set
print(f"Mean Absolute Error for the Test Set is: {mean_absolute_error(y_train**2, predictions_xg_model_train_cv**2)}")
print(f"Mean Squared Error for the Test Set is: {mean_squared_error(y_train**2, predictions_xg_model_train_cv**2)}")
print(f"Root Mean Squared Error for the Test Set is : {np.sqrt(mean_squared_error(y_train**2, predictions_xg_model_train_cv**2))}")
print(f"R squared score for the Test Set is: {r2_score(y_train**2, predictions_xg_model_train_cv**2)}")

print("__"*25 + "\n")

# Checking Metrics on Test Set
print(f"Mean Absolute Error for the Test Set is: {mean_absolute_error(y_test**2, predictions_xg_model_test_cv**2)}")
print(f"Mean Squared Error for the Test Set is: {mean_squared_error(y_test**2, predictions_xg_model_test_cv**2)}")
print(f"Root Mean Squared Error for the Test Set is : {np.sqrt(mean_squared_error(y_test**2, predictions_xg_model_test_cv**2))}")
print(f"R squared score for the Test Set is: {r2_score(y_test**2, predictions_xg_model_test_cv**2)}")
```
